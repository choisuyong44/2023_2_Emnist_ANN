## 1) Data load 및 Preprocessor

1) 그림판을 이용해서 16x16 t,u,v,w,x,y,z를 직접 손으로 그림 // character 당 100장으로 학습 및 테스트
2) PIL의 Image library를 사용하여 PNG를 array로 바꿈
3) png 파일의 이름에서 label를 가져와서 [data,label] 형태의 list로 바꿈
,나중에 학습시킬 때 배열의 첫 번째 인덱스값이 진짜 라벨
4) Train Data와 Test Data로 나눔 680 : 20<br>
여기까지는 c언어에서 처리안하고, 파이썬으로 처리할 예정

## 주의사항 :
1. Input Data는 그림판의 환경 그대로 하얀 바탕에 글자를 입력해야함. 
이후 전처리에서 255 - 로 처리함
2. PIL은 Preprocessor 과정에서만 사용했음. 혹시 없다면 pip install pillow 해야함
3. PNG 파일만 가능

```python
## PNG를 array로 바꾸는 라이브러리, 외부 라이브러리이므로 없다면 pip install pillow으로 설치해야함.
from PIL import Image
import random
import numpy as np
import matplotlib.pyplot as plt
import os


directory = '.\data'                 # input data directory 지정
dataset = []                         # list 형태로 [data,label] 받을 예정
labelnum = ['t','u','v','w','x','y','z']


# (16x16) PNG 파일을 배열로 바꾸는 함수, 0 으로 갈수록 검은색, 255로 갈수록 흰색
def imagePreprocessor(path):
    img = Image.open(path)                     # image 경로 표시
    img = img.convert('L')                     # 흑백 모드로 전환 
    input_data = 255 - np.array(img)           # 255를 빼서 바탕이 0이 되도록 만듬
    
    return input_data

    
def load_data(directory,check_data = False):                          # return 값으로 [[data1,label1],[data2,label2]] 이렇게 받을 거임.
    for filename in os.listdir(directory):         
        ## file의 label
        label = filename.replace('.png','')         # filename에서 .png 제거
        label = label[0]
        
        for i in range(len(labelnum)):
            if label == labelnum[i]:
                label = i
                break
        
        ## file의 data
        f_path = os.path.join(directory, filename) # os.path.join은 경로를 합쳐주는 메서드
        if os.path.isfile(f_path) and filename.endswith('.png'):
            f = imagePreprocessor(f_path)          # 위의 함수에서 정의 : 파일의 경로를 넣고, 16x16을 배열로 바꿈 
            
            if check_data:
                # data 확인용 [array(16x16),label]
                print(f)
                print("{}\n\n".format(label))
            

            ## 2차원 array를 flatten하게 바꾸어 줌 -> 행렬의 연산을 쉽게 하기 위해서
            ## list로 변환
            f = np.array(f,dtype=np.uint8)
            f = f.flatten()
            f = f.tolist()
            
            # dataset에 [f,label]로 넣음.
            temp = [f,label]
            dataset.append(temp)
            
            
    return dataset

def suffle_and_divide(dataset):
    random.shuffle(dataset)
    train_data, test_data = dataset[:680], dataset[680:]
    return train_data, test_data


def check_dataset(data):
    for i in data:
        print(i)


print("t->0, u->1, v->2, w->3, x->4, y->5 , z->6 로 바꿈")
dataset = load_data(directory,check_dataset)
train_data, test_data = suffle_and_divide(dataset)

# check_dataset 
# check_dataset(dataset)
# check_dataset(train_data)
# check_dataset(test_data)

# data labeling 출력 계층 label
```

# 2) 인공신경망 생성 3계층 신경망 생성


```python
import numpy as np
import matplotlib.pyplot as plt
# neural network class definition
class neuralNetwork:
    
    
    # 신경망 초기화
    def __init__(self, inputnodes, hiddennodes, outputnodes, learningrate):
        
        # 3계층으로 구성
        self.inodes = inputnodes
        self.hnodes = hiddennodes
        self.onodes = outputnodes
        
        ## 가중치 설정  : input과 hidden 사이의 가중치
        ## 가중치 설정  : hidden과 output 사이의 가중치
        self.wih = np.random.normal(0.0, pow(self.inodes, -0.5), (self.hnodes, self.inodes))
        self.who = np.random.normal(0.0, pow(self.hnodes, -0.5), (self.onodes, self.hnodes))

        # 학습률
        self.lr = learningrate
        
        # 활성화 함수
        # sigmoid overflow 방지
        self.activation_function = lambda x: 1 / (1 + np.exp(-np.clip(x, -500, 500)))

        pass

    

    # 순전파 최종적인 output값 내보냄
    def forward_propagtion(self, inputs_list):
        # np array로 바꿔야 행렬곱가능
        inputs = np.array(inputs_list,ndmin=2).T
        
        # hidden layer input값
        hidden_inputs = np.dot(self.wih, inputs)
        # hidden layer activation
        hidden_outputs = self.activation_function(hidden_inputs)
        
        # output layer input 값
        final_inputs = np.dot(self.who, hidden_outputs)
        # output layer activation(input)
        final_outputs = self.activation_function(final_inputs)
        
        return final_outputs
    
    
        # 역전파
    def train(self, inputs_list, targets_list):
        
        # 입력 리스트를 2차원 행렬로 바꾸기
        inputs = np.array(inputs_list,ndmin=2).T
        targets = np.array(targets_list,ndmin=2).T
        
        # hidden layer로 가는 신호계산
        hidden_inputs = np.dot(self.wih, inputs)
        # hidden layer * activation function
        hidden_outputs = self.activation_function(hidden_inputs)
        
        # output layer로 가는 신호 계산
        final_inputs = np.dot(self.who, hidden_outputs)
        # activation을 구해서 최종적으로 구해진 output
        final_outputs = self.activation_function(final_inputs)
        
        # output에서 error 구하기
        output_errors = targets - final_outputs
        
        # 가중치에 따라 error 나누기
        hidden_errors = np.dot(self.who.T, output_errors) 
        
        # hidden과 output layer 사이에 가중치 업데이트
        self.who += self.lr * np.dot((output_errors * final_outputs * (1.0 - final_outputs)), np.transpose(hidden_outputs))
        
        # input과 hidden layer 사이에 가중치 업데이트
        self.wih += self.lr * np.dot((hidden_errors * hidden_outputs * (1.0 - hidden_outputs)), np.transpose(inputs))
    
        pass

# inputs, hidden, output Node의 수
inputN = 256
hiddenN = 100
outputN = 10

# learning rate
lr = 0.3

# 신경망 인스턴스 생성
n = neuralNetwork(inputN,hiddenN,outputN,lr)

# epoch 설정
epoch = 5

# 학습 데이터
for e in range(epoch):
    
    for i in range(len(train_data)):
        # train data의 1번째 index가 label
        correct_label = int(train_data[i][1])
        inputs = (np.asfarray(train_data[i][0]) /255.0*0.99) + 0.01
        target = np.zeros(outputN) + 0.01
        target[train_data[i][1]] = 0.99
        n.train(inputs,target)
        pass
    pass

## 문자 환산 table
change_char = ['t','u','v','w','x','y','z']

## 정답을 몇개를 맞췄는지 알 수 있도록 점수
score = []
correct_count = 0
for i in range(len(test_data)):
    correct =  change_char[int(test_data[i][1])]
    outputs = n.forward_propagtion(test_data[i][0])
    predict = change_char[np.argmax(outputs)]
    
    print("\n")
    if(correct == predict):
        print("정답입니다.",end =" | ")
        score.append(1)
        correct_count += 1
    else:
        print("오답입니다.",end =" | ")
        score.append(0)
    print("target = ",correct,", ",end=" ")
    print("predict = ",predict)

        
print("\n\n정답률 :",correct_count/20 * 100)
```
